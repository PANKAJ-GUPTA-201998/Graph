```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/RESOURCE-ALL-FOUR.csv')
```

```python
df.head()
```

```python
a = df['INDEX(MINUTES)']
```

```python
b = df['IQRMC']
```

```python
c = df['IQRMT']
```

```python
d = df['LRMMT']
```

```python
e = df['TD4VM']
```

```python

plt.figure(figsize=(12,6))
plt.plot(a,d,color='r',linewidth='0.5')
plt.plot(a,b,color='k',linewidth='0.5')
plt.plot(a,e,color='m',linewidth='0.75')
plt.plot(a,c,color='y',linewidth='0.5')
plt.xlabel('Time Minutes')
plt.ylabel('VM Resources Utilization')
plt.title('Comparison of VM Utilization of TD4VM with other Models')
plt.legend(['LRMMT','IQRMC','TD4VM','IQRMMT'])
plt.show()
```
